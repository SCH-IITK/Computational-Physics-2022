## Real pendulum
Till now we have solved the equation of pendulum $ \ddot\theta = - \frac{g}{L} \sin(\theta) $ by applying the small angle approximation $\sin(\theta) \approx \theta$. This is because we can't solve the original equation analytically without small angle approximation, but after learning the techniques in this course we can find numerical solutions.

1) Find the solution of the real pendulum and plot its angular displacement as a function of time. Compare it with the trajectory obtained using small angle approximation. Do the comparison for small initial angles(approximately 10-30$^\circ$) as well as for large initial angles(approximately 90$^\circ$).

2) To make it more realistic add a damping term to the above equation. Plot and observe the solution for different initial conditions and different values of damping coefficient.
$$ \ddot\theta = - \frac{g}{L} \sin(\theta) - \beta\dot\theta$$

3) As the final step add a periodic external force, and again observe the solution for different values of the parameters.
$$ \ddot\theta = - \frac{g}{L} \sin(\theta) - \beta\dot\theta + F\sin(\omega t)$$

## Schrodinger equation
The Schrödinger equation is a linear partial differential equation that governs the wave function of a quantum-mechanical system. Wave function of a particle or a system contains the information of the particle and the system. The Schrödinger equation is a very important equation and is the heart of quantum mechanics.

In some simple cases we can ignore the time dependence of the Schrödinger equation. Considering only 1D the time independent Schrödinger equation is given by 
$$ -\frac{\hbar ^2}{2m} \frac{d^2\psi}{dx^2} + V(x)\psi = E\psi$$
where $\psi(x)$ is the wavefunction.

https://en.wikipedia.org/wiki/Schr%C3%B6dinger_equation

1) Plot and find the wavefunction for a particle in 1D box (1D infinite potential well) numerically. Try to guess the functional form from the numeric solution.

2) Plot and find the wavefunction for a delta potential and double delta potential.

Note that the wavefunction must be continuous at all points and the integral of $\psi$ from $\infty$ to $-\infty$ should be finite (which means that $\psi$ should tend to 0 at infinity for both the above cases! Try to figure out why). Observe that valid solutions for $\psi$ may not exist for all values of $E$, which gives rise to quantisation of energy!
