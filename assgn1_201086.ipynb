```python
#Code 3.a

import numpy as np
def MagF(z,t):
    print("-E0/c[",2*np.sin(z-t),"x̂ + ", np.cos(z-t), "ŷ]")
    
for z in np.arange(0.0, 4*np.pi, 0.4*np.pi):
    for t in np.arange(0.0, 10.0, 1.0):
        MagF(z,t)
```

```python
#Code 3.b

def Pyt(z,t):
    print("-E0^2/c[", 1 + 3*(np.sin(z-t))**2, "]ẑ")
    
for z in np.arange(0.0, 4*np.pi, 0.4*np.pi):
    for t in np.arange(0.0, 10.0, 1.0):
        Pyt(z,t)
```

```python
#Code1

import matplotlib.pyplot as plt

t = np.arange(0, 2*(np.pi), 0.01) 
x = 10*np.cos(t)*(np.sin(t) + (1 + (np.sin(t))**2)**0.5) 

idx = np.argwhere(np.diff(np.sign(x - 11.5))).flatten()
print ("The theta's at which x = 11.5 are ", t[idx])
plt.plot(t,x)
plt.show()
```

```python
#Code2

def f(x, y,nx,ny):
  return (((2/3)**0.5)/np.pi)*np.sin(nx*x/2)*np.sin(ny*y/3)

xlist = np.linspace(0, 2*np.pi, 100)
ylist = np.linspace(0, 3*np.pi, 100)
X, Y = np.meshgrid(xlist, ylist)

def PlotGraph(nx, ny):

    #creating 3D plane
    Z = f(X, Y, nx, ny)

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(projection='3d')
    #ax.set_title('Surface Plot')

    #drawing surface plot
    surface_plot= ax.plot_surface(X, Y, Z, cmap='viridis')

    plt.show()
  
print("How many graphs do you want to plot?")
n = int(input())
i = 0
while i<n:
    print("insert values of nx and ny")
    PlotGraph(int(input()), int(input()))
    i = i+1
```
