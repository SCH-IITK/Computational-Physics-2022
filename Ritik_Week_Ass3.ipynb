```python
def f(Eb):
    return np.sqrt(80-Eb)*np.tan(np.sqrt(80-Eb))-np.sqrt(Eb)
Eb=np.linspace(0,80,200)
plt.figure(figsize=(10,10))
plt.grid()
plt.plot(Eb,f(Eb))

#from above it is clear that it have three roots
#Bisection_Method

#1st_Root
a1 = 59
b1 = 70
e = 0.00001
n1 = 0
x1 = (a1+b1)/2

while(abs(f(x1))>e and n1<10000):
    if(f(x1)*f(a1)<0) :
        b1=x1
    else:
        a1=x1
    x1=(a1+b1)/2.0
    n1=n1+1
    
print(x1,f(x1))
print(n1)

#2nd_Root
a2 = 20
b2 = 50
e = 0.00001
n2 = 0
x2 = (a2+b2)/2

while(abs(f(x2))>e and n2<10000):
    if(f(x2)*f(a2)<0) :
        b2=x2
    else:
        a2=x2
    x2=(a2+b2)/2.0
    n2=n2+1
    
print(x2,f(x2))
print(n2)

#3rd_Root
a3 = 78
b3 = 80
e = 0.00001
n3 = 0
x3 = (a3+b3)/2


while(abs(f(x3))>e and n3<10000):
    if(f(x3)*f(a3)<0) :
        b3=x3
    else:
        a3=x3
    x3=(a3+b3)/2.0
    n3=n3+1
    
print(x3,f(x3))
print(n3)



```

```python
#Newton-Raphson method
def f1(x):
    h=0.00001
    return (f(x+h)-f(x-h))/(2*h)

#R00t_1
e=0.00001
x=40
n1=0
while((abs(f(x))>e) and (n1<10000)):
    x1=x-f(x)/f1(x)
    x=x1
    n1=n1+1
print(x,f(x))
print(n1)

#Root_2
e=0.00001
x=60
n2=0
while((abs(f(x))>e) and (n2<10000)):
    x1=x-f(x)/f1(x)
    x=x1
    n2=n2+1
print(x,f(x))
print(n2)

#Root_3
e=0.00001
x=78
n3=0
while((abs(f(x))>e) and (n3<10000)):
    x1=x-f(x)/f1(x)
    x=x1
    n3=n3+1
print(x,f(x))
print(n3)

```
