```python
#numpy
#Q1
import numpy as np
def revarray(a,n):
    if(n%2==0):
        for i in range(int(n/2)):
            t=a[n-i-1]
            a[n-i-1]=a[i]
            a[i]=t
    else:
        for i in range(int((n-1)/2)):
            t=a[n-i-1]
            a[n-i-1]=a[i]
            a[i]=t
#checking
a=[1,2,3,4,5]
revarray(a,5)
print(a)
```

```python
#Q2-determinant of a 3D matrix
def det(a):
    return (a[0][0]*(a[1][1]*a[2][2]-a[1][2]*a[2][1])-a[0][1]*(a[1][0]*a[2][2]-a[1][2]*a[2][0])+a[0][2]*(a[1][0]*a[2][1]-a[1][1]*a[2][0]))
#checking
a=[[2,0,0],[0,2,0],[0,0,2]]
det(a)
```

```python
#Q3
#A
#given-E(z,t)=E0cos(z-t)x^+2E0cos(z-t+pi/2)y^
#we know-B(z,t)=(1/c)(z^crossE(z,t))
#Let E0=1
import numpy as np
z=np.linspace(0,np.pi*4,100)    
t=np.linspace(0,10,100)
Z,T=np.meshgrid(z,t)
e=1 
c=3*10**8
u=4*np.pi*10**(-7)
def ycomE(z,t):
    return (2)*e*np.cos(z-t+np.pi/2)
def xcomE(z,t):
    return (e)*np.cos(z-t)
def zcomE(z,t):
    return 0*z*t
def xcomB(z,t):
    return -(2/c)*e*np.cos(z-t+np.pi/2)
def ycomB(z,t):
    return (e/c)*np.cos(z-t)
def zcomB(z,t):
    return 0*z*t

B=np.array([xcomB(Z,T),ycomB(Z,T),zcomB(Z,T)])
E=np.array([xcomE(Z,T),ycomE(Z,T),zcomE(Z,T)])
print(B)
```

```python
#mathplotlib
#Q1
#Vo=10,yo=5m, go=10m/s^2 x=11.5
#y=xtanA-(1/2)*g*x**2/(vo**2cos**2(A))=-5
#x=(Vo*sin(A)+root(sin**2(A)*Vo**2-2*g*yo))*Vo*cos(A)*(1/g)
import numpy as np
import matplotlib.pyplot as plt
import math as m
Vo=10
yo=-5
g=10
def f(x):
    return (Vo*m.sin(x)+np.sqrt(((np.sin(x))**2)*Vo**2-2*g*yo))*Vo*np.cos(x)*(1/g)
x=np.linspace(0,2,1000)
y=np.linspace(0,0,1000)
for i in range(1000):
   y[i]=f(x[i])
plt.figure(figsize=(12,7))
plt.plot(x,y)
plt.plot([0,1],[11.5,11.5],'c--')
plt.xlabel('theta',size=20)
plt.ylabel('x',size=20)
plt.grid()
plt.show()
p=np.linspace(0.15,0.17,1000)
q=np.linspace(0,0,1000)
for i in range(1000):
    q[i]=f(p[i])
plt.figure(figsize=(12,7))
plt.plot(p,q)
plt.xlabel('theta',size=20)
plt.ylabel('x',size=20)
plt.grid()
plt.show()
print('guess from graph for theta is 0.1525')
```

```python
#Matplotlib
#Q2
#
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
Lx=2*np.pi
Ly=3*np.pi
P=np.pi
def si(nx,ny,x,y):
     return (2/(np.sqrt(Lx*Ly)))*np.sin((nx*P*x)/(Lx))*np.sin((ny*P*y)/(Ly))
x=np.linspace(-5*P,5*P,101)
y=np.linspace(-5*P,5*P,101)
X,Y=np.meshgrid(x,y)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
#print(np.shape(X),np.shape(Y),np.shape(si(1,1,X,Y)))
ax.plot_surface(X,Y,si(1,1,X,Y),cmap='plasma')
plt.show
```
