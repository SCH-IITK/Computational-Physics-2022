```python
# question 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def odes(x,t):
    #constants
    g=9.81
    l=1
    
    theta_1=x[0]
    theta_2=x[1]
    dtheta_1dt=theta_2
    dtheta_2dt=(-1)*g*np.sin(theta_1)/l
    
    return [dtheta_1dt,dtheta_2dt]

def odetheta(y,t):
    #constants
    g=9.81
    l=1
    
    theta_1s=y[0]
    theta_2s=y[1]
    dtheta_1sdt=theta_2s
    dtheta_2sdt=(-1)*g*(theta_1s)/l
    
    return [dtheta_1sdt,dtheta_2sdt]
    


# initial conditions small initial angle: 20 degree
x0=[np.pi/9,0]

t=np.linspace(0,10,140)

x=odeint(odes,x0,t)
y=odeint(odetheta,x0,t)

theta_1=x[:,0]
theta_2=x[:,1]

theta_1s=y[:,0]
theta_2s=y[:,1]
plt.figure(figsize=(12,7))
plt.plot(t,theta_1,label='no small angle approx')
plt.plot(t,theta_1s,label='small angle approx')

plt.legend()
plt.grid()
plt.show()

    
```

```python
def odes(x,t,F,w):
    #constants
    g=9.81
    l=1
    b1=0.09
    theta_1=x[0]
    theta_2=x[1]
    dtheta_1dt=theta_2
    dtheta_2dt=((-1)*g*np.sin(theta_1)/l)-(b1*theta_2)+F*np.sin(w*t)
    
    
    return [dtheta_1dt,dtheta_2dt]

# initial conditions 
x0=[np.pi/9,0]
F_1=2
w_1=2
F_2=3
w_2=4
F_3=6
w_3=5
t=np.linspace(0,10,140)

x=odeint(odes,x0,t,args=(F_1,w_1))
theta_1=x[:,0]
y=odeint(odes,x0,t,args=(F_2,w_2))
theta_2=y[:,0]
z=odeint(odes,x0,t,args=(F_3,w_3))
theta_3=z[:,0]

plt.figure(figsize=(12,7))
plt.plot(t,theta_1,label='F1 W1')
plt.plot(t,theta_2,label='F2 W2')
plt.plot(t,theta_3,label='F3 W3')

plt.legend()
plt.grid()
plt.show()
```
