```python
import matplotlib.pyplot as plt
import numpy as np
import math
def f(x, y):
  return np.cos(x)+2*np.sin(y)

xcord=np.linspace(0,2*math.pi, 100)
ycord=np.linspace(0,3*math.pi, 100)
X, Y = np.meshgrid(xcord,ycord)
for nx in range(3):
    for ny in range(3):
        psi=(2/(math.sqrt(6*math.pi)))*np.sin(0.5*nx*X)*np.sin((1/3)*ny*Y)

        fig = plt.figure()
        ax = fig.add_subplot(projection='3d')
        ax.set_title('Surface Plot')
        plt.title('Nx='+str(nx)+'& Ny='+str(ny))
        surface_plot = ax.plot_surface(X, Y, psi)
        plt.show()
```
