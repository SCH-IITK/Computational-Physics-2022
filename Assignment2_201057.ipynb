1. Second-order forward difference scheme


$$f_{i+1} = f_i + \Delta x f'_i + \frac{\Delta x^2}{2!}f''_i + \frac{\Delta x^3}{3!}f'''_i + ...$$
$$f_{i+2} = f_i + 2\Delta x f'_i + \frac{4\Delta x^2}{2!}f''_i + \frac{8\Delta x^3}{3!}f'''_i + ...$$
$$f_{i+1} - f_i = \Delta x f'_i + \frac{\Delta x^2}{2!}f''_i + \frac{\Delta x^3}{3!}f'''_i + ...     -(1)$$        
$$f_{i+2} - f_i = 2\Delta x f'_i + \frac{4\Delta x^2}{2!}f''_i + \frac{8\Delta x^3}{3!}f'''_i + ...      -(2)$$
4(1)-(2)
$$ 4f_{i+1} - 3f_i +f_{i+2} =  2\Delta x f'_i - \frac{4\Delta x^3}{3!}f'''_i$$
divide by $2\Delta x$
$$f'_{i} \approx \frac{4f_{i+1}-3f_{i}-f_{i+2}}{2 \Delta x}$$

```python
#exact solution of ques1 in matplotlib using root solving method
import numpy as np
import matplotlib.pyplot as plt 

#defining the function here x is theta and f(x) is distance in meter

def f(x):
    return 10*np.cos(x)*(np.sin(x)+np.sqrt(np.sin(x)*np.sin(x)+1))-11.5   #at x=11.5 m functions gives 0
x = np.linspace(0,np.pi/2,100)
#plotting the function so we can make approximate guess from the graph
plt.plot(x,f(x))
plt.xlabel('angle in radians')
plt.ylabel('distance in metres')
plt.grid()
plt.show()

#from the graph we can see that f(x) has opposite signs at 0.1 and 0.2    (x=0.15 approx)
a=0.1
b=0.2
#initial guesses

error=0.01
#error tolerance

x=(a+b)/2


#bisection method algorithm
while(abs(f(x))>error):
    if(f(x)*f(a)<0):
        b=x
    else:
        a=x
    x=(a+b)/2.0
    
    
print(x,"radians ")

#from the graph we can see that f(x) has opposite signs at 0.9 and 1.1    (x=1 approx)
a=0.9
b=1.1
#initial guesses

error=0.01
#error tolerance

x=(a+b)/2


#bisection method algorithm
while(abs(f(x))>error):
    if(f(x)*f(a)<0):
        b=x
    else:
        a=x
    x=(a+b)/2.0
    
    
print(x,"radians ")
```

```python
# integral of exp(-x^2)
import numpy as np
def f(x):
    return 2*np.exp((-1)*x*x)   # exp(-x^2) is an even function so integral will get double after calculating for positive axis 

x=np.linspace(0,1000,100001) 

mid=0

for i in range(100000):
    mid=mid+f((x[i]+x[i+1])/2)

mid=mid*1000/100000

print(mid)
```
