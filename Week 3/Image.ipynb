# Images

`imshow()` function is used in python to display data as an image.
https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.imshow.html

```python
import numpy as np
import matplotlib.pyplot as plt

img=np.random.rand(50,50) #generating a random array to be displayed as the image
print(img)
```

```python
plt.imshow(img)
#making a 2d image from the array
```

We can use vmax and vmin arguments to specify the maximum and minimum intensity.

```python
plt.imshow(img,vmax=0.8,vmin=.2)
plt.colorbar(shrink=0.5)
```

We can plot a histogram of the image to see the number of points with a given intensity. To plot the histogram, we need to convert the 2D array into 1D using `np.flatten()`

```python
histogram = plt.hist(img.flatten(), bins='auto')
```

As we can see the intensity is distributed uniformly throughout, which is what we expect from a random number generator.

```python
plt.imshow(img,cmap='bone')
plt.colorbar()
plt.show()
plt.imshow(img,cmap='plasma')
plt.colorbar()
plt.show()
plt.imshow(img,cmap='Purples')
plt.colorbar()
plt.show()
plt.imshow(img,cmap='RdBu')
plt.colorbar()
plt.show()
```

We can sort the img data to get a gradient image

```python
print(img)
img=np.sort(img,axis=0)
plt.imshow(img)
plt.show()
img=np.sort(img,axis=1)
plt.imshow(img)
```

## Image layering
We can layer two images using imshow()

```python
#lets plot a wave generated by a source

def f(x,y):
    return np.sin(x**2+y**2)

x1=np.linspace(-5,5,100)
y1=np.linspace(-5,5,100)

X,Y=np.meshgrid(x1,y1)
Z=f(X,Y)

plt.imshow(Z,cmap='summer')
plt.colorbar()
```

```python
extent = np.min(x1), np.max(x1), np.min(y1), np.max(y1)
#making sure that both images have same shape 

plt.imshow(Z,cmap='summer',extent=extent)

plt.imshow(img,cmap='Purples',alpha=0.7,extent=extent)
#here alpha is used set the transparency of the second image
```

## Image stacking
Image stacking is a method to improve the quality of a noisy image by 'stacking' several such images on top of each other. This helps in cancelling out some random noise which may be present in the image and improves the image quality.

```python
z1=Z+2*np.random.rand(100,100) #adding random noise to the image
plt.imshow(z1,cmap='winter')
```

```python
z2=Z+2*np.random.rand(100,100)
z3=Z+2*np.random.rand(100,100)
z4=Z+2*np.random.rand(100,100)
z5=Z+2*np.random.rand(100,100)
plt.imshow(z1+z2+z3+z4+z5,cmap='winter')
#We can see how the image quality improves on stacking
```

## Reading an image from a file
`imread()` function is used to read an image from a file on your computer. Let import `Diffraction_ss_z=200.jpg`, which is an image of the pattern obtained on the screen from single slit diffraction. This image is from an experiment which I did in PHY224(optics) lab, which some of you will also be doing in future.

Single slit diffraction theory:https://phys.libretexts.org/Bookshelves/University_Physics/Book%3A_University_Physics_(OpenStax)/University_Physics_III_-_Optics_and_Modern_Physics_(OpenStax)/04%3A_Diffraction/4.02%3A_Single-Slit_Diffraction

```python
import matplotlib.image as mpimg
img = mpimg.imread('Diffraction_ss_z=200.jpg')
plt.imshow(img)
```

```python
print(img)
#here the image is a 3D array as it contains RGB values at each position

d1=np.sum(img,axis=2)
#d1 is a 2D array which just denotes the intensity at each point

#lets plot the intensity variation against x
d2=np.sum(d1,axis=0)
print(d1)
plt.plot(d2)
```

The above graph is quite similar to what we expect theoretically, though there are some errors like minima not going to 0. Most of such errors due to the imperfections in the camera which captured this image.

```python
plt.rcParams["figure.figsize"] = [10.00, 5]
plt.imshow(d1)
plt.colorbar(shrink=0.5)
plt.show()

#lets try to increase the contrast of the image around the central maxima

d1=d1/70
d1=np.exp(d1)
plt.rcParams["figure.figsize"] = [10.00, 5]
plt.imshow(d1,vmin=0,vmax=40000)
plt.colorbar(shrink=0.5)
```

We can see that the image contrast is much better for high intensity points.
