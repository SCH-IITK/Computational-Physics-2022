```python
#Question 1

import numpy as np
import matplotlib.pyplot as plt
filename = 'velocities.txt'
d = np.loadtxt(filename)
I =0 
dis = []
for i in range(len(d)-1):
    I = I + abs(((d[i][1]+d[i+1][1])/2)*(d[i+1][0]-d[i][0]))
    dis.append(I) 
    
dis.append(I)    
print("Total distance travelled = ",I) 
plt.plot(d[:,0],d[:,1])
plt.plot(d[:,0],dis)
plt.show()
plt.plot(d[:,0],d[:,1])
plt.show()
```
