```python
import numpy as np
import matplotlib.pyplot as plt
V=80
Eb=np.linspace(0,80,1001)
LHS=(np.sqrt(V-Eb)*np.tan(np.sqrt(V-Eb)))
RHS=np.sqrt(Eb)
plt.figure(figsize=(12,12))
plt.plot(Eb,LHS-RHS)
plt.grid()
plt.show()
```

```python
#3
#Newton Raphon's algorithm
import math as m
def f1(x):
    return((-1/(2*np.sqrt(V-x)))*np.tan(np.sqrt(V-x))-(1/2)*(1/(m.cos(np.sqrt(V-x))))**2-1/(2*np.sqrt(x)))
```

```python
error=0.000001
x=30
n1=0
while ((f(x)>error) and (n1<10000)):
    x=x-f(x)/f1(x)
    n1=n1+1
print('First solution is '+str(x))
```
