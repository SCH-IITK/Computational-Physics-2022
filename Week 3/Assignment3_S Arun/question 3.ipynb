```python
import numpy as np
st=np.loadtxt('stm.txt')
print(st)
print(st.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(st)
plt.show()
```

```python
from mpl_toolkits import mplot3d
y=np.linspace(0,662,663)
x=np.linspace(0,675,676)
X,Y=np.meshgrid(x,y)
Z=st
print(X.shape,Y.shape,Z.shape)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
plot=ax.plot_surface(X,Y,Z,cmap='viridis')
plt.colorbar(plot,shrink=0.5)
plt.show()
```
