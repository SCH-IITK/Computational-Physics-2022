```python
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.linspace(0,math.pi/2,num=100)
y=[0]*100
for i in range(0,100):
    y[i]=(10*np.sin(x[i])*np.cos(x[i]) + np.cos(x[i])*math.sqrt((100*np.sin(x[i])*np.sin(x[i]))+50))

plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Range')
plt.show()
```
