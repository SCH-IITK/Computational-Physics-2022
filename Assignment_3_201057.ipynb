```python
#question1#
import matplotlib.pyplot as plt

import numpy as np

textfile='velocities.txt'
arr2D=np.loadtxt(textfile)
y1=np.array(arr2D[:,1])
x=np.array(arr2D[:,0])
plt.plot(x,y1)
mylist=[0]
I=0
for i in range(0,100):
    I=I+abs((y1[i]+y1[i+1])/2)
    mylist.append(I)
y2=np.array(mylist)
plt.figure(figsize=(7,7))
plt.plot(x,y1,label='velocity')
plt.plot(x,y2,label='distance')

plt.legend()

plt.xlabel('time',size=20)
plt.ylabel('Y Axis',size=20)

plt.grid()
plt.show()

 
```

```python
#question2#


#defining the equation with even parity
def f(x):
    return np.sqrt(80-x)*np.tan(np.sqrt(80-x))-np.sqrt(x)


#plotting the function so we can make approximate guess from the graph
x=np.linspace(0,80,1000)
plt.plot(x,f(x))

plt.ylim(-50 , 50)
#plt.xlim(77.5 , 79)

plt.grid()
plt.show()
```

```python

a=20      # solution 1
b=40
#initial guesses

error=0.01
#error tolerance

x=(a+b)/2


#bisection method algorithm
while(abs(f(x))>error):
    if(f(x)*f(a)<0):
        b=x
    else:
        a=x
    x=(a+b)/2.0
    
    
sol_even=[]
sol_even.append(x)

a=60        # solution 2
b=65
#initial guesses

error=0.01
#error tolerance

x=(a+b)/2.0


#bisection method algorithm
while(abs(f(x))>error):
    if(f(x)*f(a)<0):
        b=x
    else:
        a=x
    x=(a+b)/2.0
    
    
sol_even.append(x)

a=77.7          #solution 3
b=79
#initial guesses

error=0.001
#error tolerance

x=(a+b)/2.0


#bisection method algorithm
while(abs(f(x))>error):
    if(f(x)*f(a)<0):
        b=x
    else:
        a=x
    x=(a+b)/2.0
    
    
sol_even.append(x)

print(sol_even)
```

```python
#defining the equation with odd parity
def f(x):
    return np.sqrt(80-x)*np.reciprocal(np.tan(np.sqrt(80-x)))-np.sqrt(x)


#plotting the function so we can make approximate guess from the graph
x=np.linspace(0,80,1000)
plt.plot(x,f(x))

plt.ylim(-50 , 50)
#plt.xlim(65 , 69)

plt.grid()
plt.show()
```

```python
# newton raphson method to campare some solution and speed of the two methods

#defining the derivative of the equation with odd parity
def f1(x):
    return np.reciprocal(np.square(np.sin(np.sqrt(80-x))))/2 - np.reciprocal(np.sqrt(80-x)*np.tan(np.sqrt(80-x)))/2 - np.reciprocal(np.sqrt(x))/2
n1=0        #solution 1
error=0.01
x=25

while((abs(f(x))>error) and (n1<100)):
    x1=x-f(x)/f1(x)
    x=x1
    n1=n1+1

print(x,f(x))

n1=0         # solution 2
error=0.01
x=65

while((abs(f(x))>error) and (n1<100)):
    x1=x-f(x)/f1(x)
    x=x1
    n1=n1+1

print(x,f(x))


```

newton raphson method is faster as it takes lesser number of iteration
