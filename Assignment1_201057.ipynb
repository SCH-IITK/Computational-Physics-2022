1. Write a NumPy program to reverse an array (first element becomes last).

```python
mylist=[]

n = int(input())
for i in range(0, n):
    ele = int(input())
              
    mylist.append(ele)
              
import numpy as np

arr =np.array(mylist)
print(arr[::-1])
```

 1.Write a NumPy program for finding the determinant of the 3D input matrix.

```python
matrix=[]
for i in range(3):
    row=[]
    for j in range(3):
        row.append(int(input()))
    matrix.append(row)
    
import numpy as np

arr=np.array(matrix)
x=arr[0][0]*(arr[1][1]*arr[2][2]-arr[1][2]*arr[2][1])-arr[0][1]*(arr[1][0]*arr[2][2]-arr[1][2]*arr[2][0])+arr[0][2]*(arr[1][0]*arr[2][1]-arr[1][1]*arr[2][0])
print(x)
```

Find magnetic field and poynting vector for a given electric field

```python
import numpy as np 

Eo=int(input())
z=np.linspace(0,4*np.pi,100)
time=int(input())
t=np.full(100,time)

a=2*Eo*np.sin(z-t)
b=Eo*np.cos(z-t)
print('x component of magnetic field:a')
print('y component of magnetic field:b')
c=Eo*Eo*(1+3*np.sin(z-t)*np.sin(z-t))
print('c is poynting vector')
uo=4*22/7*(10**(-7))
for i in range(100):
    print(str(a[i])+' x +'+str(b[i])+' y')
    print(str(c[i]/uo)+' z'+' poynting vector')

    
```

Find value of theta for a given value of x=11.5m

```python
import matplotlib.pyplot as plt

import numpy as np 

theta = np.linspace(0,np.pi/2,100)
t=np.sin(theta)*np.sin(theta)+1
x=10*np.cos(theta)*(np.sin(theta)+np.sqrt(t))
plt.plot(theta,x)
plt.xlabel('angle in radians')
plt.ylabel('x in metres')
plt.grid()
plt.show()

```

2.Wavefunction for different values of nx and ny

```python
import matplotlib.pyplot as plt

import numpy as np 
from mpl_toolkits import mplot3d

xlist = np.linspace(0, 2*np.pi, 100)
ylist = np.linspace(0, 3*np.pi, 100)
X, Y = np.meshgrid(xlist, ylist)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2, 2, 1, projection='3d')
Z=((2/3)**0.5)*np.sin(1*X/2)*np.sin(1*Y/3)/np.pi # nx=1 and ny=1
ax.plot_surface(X, Y, Z,cmap='viridis')

ax = fig.add_subplot(2, 2, 2, projection='3d')
Z=((2/3)**0.5)*np.sin(2*X/2)*np.sin(1*Y/3)/np.pi# nx=2 and ny=1
ax.plot_surface(X, Y, Z,cmap='viridis')

ax = fig.add_subplot(2, 2, 3, projection='3d')
Z=((2/3)**0.5)*np.sin(2*X/2)*np.sin(2*Y/3)/np.pi # nx=2 and ny=2
ax.plot_surface(X, Y, Z,cmap='viridis')

ax = fig.add_subplot(2, 2, 4, projection='3d')
Z=((2/3)**0.5)*np.sin(2*X/2)*np.sin(3*Y/3)/np.pi # nx=2 and ny=3
ax.plot_surface(X, Y, Z,cmap='viridis')

plt.show()
```
