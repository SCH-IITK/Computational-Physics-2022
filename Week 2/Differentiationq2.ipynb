```python
import matplotlib.pyplot as plt
import numpy as np
import math

fwd=[0]*1000
ctrl=[0]*1000
er1=[0]*1000
er2=[0]*1000

x=np.array([0.1,1,100])
h=np.linspace(0.001,1,1000)

for j in x:
    for i in range(1000):
        fwd[i]=(np.cos(j+h[i])-np.cos(j))/h[i]
        ctrl[i]=(np.cos(j+h[i])-np.cos(j-h[i]))/(2*h[i])
        er1[i]=abs(-np.sin(j)-fwd[i])
        er2[i]=abs(-np.sin(j)-ctrl[i])

    plt.figure(figsize=(10,10))
    plt.subplot(2,2,1)
    plt.plot(h,fwd,'c-')
    plt.xlabel('h',size=10)
    plt.ylabel('Forward derivative',size=10)
    plt.title('Derivative of cos function\n'+'t='+str(j),size=10)
    print("                              ")
    print("\n\n")

    plt.subplot(2,2,3)
    plt.plot(h,er1,'r-')
    plt.xlabel('h',size=10)
    plt.ylabel('Error',size=10)
    plt.title('t='+str(j),size=10)
    print("                              ")
    
    plt.subplot(2,2,2)
    plt.plot(h,ctrl,'c-')
    plt.xlabel('h',size=10)
    plt.ylabel('Central Derivative',size=10)
    plt.title('Derivative of cos function\n'+'t='+str(j),size=10)
    print("\n\n")

    plt.subplot(2,2,4)
    plt.plot(h,er2,'r-')
    plt.xlabel('h',size=10)
    plt.ylabel('Error',size=10)
    plt.title('t='+str(j),size=10)
    
    plt.subplots_adjust(
                    wspace=0.4, 
                    hspace=0.4)
    
    plt.show()
    
for j in x:
    for i in range(1000):
        fwd[i]=(math.exp(j+h[i])-math.exp(j))/h[i]
        ctrl[i]=(math.exp(j+h[i])-math.exp(j-h[i]))/(2*h[i])
        er1[i]=abs(math.exp(j)-fwd[i])
        er2[i]=abs(math.exp(j)-ctrl[i])

    plt.figure(figsize=(10,10))
    plt.subplot(2,2,1)
    plt.plot(h,fwd,'c-')
    plt.xlabel('h',size=10)
    plt.ylabel('Forward derivative',size=10)
    plt.title('Derivative of exp function\n'+'t='+str(j),size=10)
    print("                              ")
    print("\n\n")

    plt.subplot(2,2,3)
    plt.plot(h,er1,'r-')
    plt.xlabel('h',size=10)
    plt.ylabel('Error',size=10)
    plt.title('t='+str(j),size=10)
    print("                              ")

    plt.subplot(2,2,2)
    plt.plot(h,ctrl,'c-')
    plt.xlabel('h',size=10)
    plt.ylabel('Central Derivative',size=10)
    plt.title('Derivative of exp function\n'+'t='+str(j),size=10)
    print("\n\n")

    plt.subplot(2,2,4)
    plt.plot(h,er2,'r-')
    plt.xlabel('h',size=10)
    plt.ylabel('Error',size=10)
    plt.title('t='+str(j),size=10)

    plt.subplots_adjust(
                    wspace=0.4, 
                    hspace=0.4)

    plt.show()



```

When h is very small, the value of the derivative is almost equal to the slope of the tangent of the function at the given point. The error in the computed value of derivative is approximately zero.
